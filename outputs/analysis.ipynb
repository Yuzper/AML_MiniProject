```python
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = "/home/aver/Machine_Learning/AML_MiniProject/outputs/metrics/roberta_full_history.json"
with open(file_path, "r") as file:
    data = json.load(file)
```

```python
# Load data
with open('/home/aver/Machine_Learning/AML_MiniProject/outputs/metrics/bert_local_run.json') as f:
    data = json.load(f)

df = pd.DataFrame(data)

# Extract subsets
train_df = df[df['loss'].notna()]
eval_df = df[df['eval_loss'].notna()]

# Smooth training loss (optional)
train_df['loss_smooth'] = train_df['loss'].rolling(window=10).mean()

# Create plot
fig, ax1 = plt.subplots()

# Plot training + eval loss on left axis
ax1.plot(train_df['step'], train_df['loss_smooth'], label='Train Loss (smoothed)', color='orange')
ax1.scatter(eval_df['step'], eval_df['eval_loss'], label='Eval Loss', color='red')
ax1.set_xlabel('Step')
ax1.set_ylabel('Loss', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')

# Plot eval accuracy on right axis
ax2 = ax1.twinx()
ax2.scatter(eval_df['step'], eval_df['eval_accuracy'], label='Eval Accuracy', color='blue')
ax2.set_ylabel('Accuracy', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.set_ylim(0, 1)  # accuracy between 0 and 1

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.title('Training Loss, Eval Loss, and Eval Accuracy over Steps')
plt.grid(True)
plt.show()
```
